```python
import numpy as np  
import matplotlib.pyplot as plt
```

```python
'''Resolvendo o exercicio 1: Força pontual em uma barra engastada'''
def ex1(intensidadeForca, comprimento):
    momento = intensidadeForca * comprimento

    print("Os esforços reativos no engastamento são:")
    print("\nForça horizontal reativa: 0")
    print("\nForça vertical reativa: ", intensidadeForca)
    print("\nMomento reativo: ", momento)
    print("Força reativa no sentido positivo do eixo y")

    #Função Normal
    @np.vectorize
    def N(x):
        return 0

    t1 = np.linspace(0.0, comprimento)
    plt.plot(t1, N(t1), color='xkcd:sky blue')
    plt.fill_between(t1, N(t1), 0, alpha=0.2, color='xkcd:sky blue')
    plt.title('Força Normal')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('N(x)') #definindo nome do eixo Y
    plt.grid()
    plt.show()
    
    #Função Cortante
    @np.vectorize
    def V(x):
      return intensidadeForca

    x = np.linspace(0, comprimento)
    plt.plot(x, V(x), color='#f55a50')
    plt.fill_between(x, V(x), 0, alpha=0.2, color='#f55a50')
    plt.title('Força Cortante')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('V(x)') #definindo nome do eixo Y
    plt.grid()
    plt.show()
    
    #Função Momento Fletor
    x = np.linspace(0,comprimento)
    def M(x):
      return (intensidadeForca)*x - momento

    plt.plot(x,M(x), color='m')
    plt.fill_between(x, M(x), 0, alpha=0.2, color='m')
    plt.title('Momento Fletor')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('M(x)') #definindo nome do eixo Y
    plt.gca().invert_yaxis()
    plt.grid()
    plt.show()

'''Resolvendo o exercicio 2: Força contínua retangular em uma barra engastada'''
def ex2(intensidadeForca, comprimento):
    #Esforços reativos
    print("Os esforços reativos no engastamento são:")
    print("\nForça horizontal reativa: 0")
    print("\nForça vertical reativa: ", intensidadeForca*comprimento)
    print("\nMomento reativo: ", intensidadeForca/2*comprimento**2)

    #Função Força Normal
    @np.vectorize
    def N(x):
        return 0

    x = np.linspace(0, comprimento)
    plt.title('Força Normal')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('N(x)') #definindo nome do eixo Y
    plt.plot(x, N(x), color="xkcd:sky blue")
    plt.grid()
    plt.show()
    print('\n')

    #Função Força Cortante
    def V(x):
        return - intensidadeForca * x + intensidadeForca * comprimento

    x = np.linspace(0, comprimento)
    plt.title('Força Cortante')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('V(x)') #definindo nome do eixo Y
    plt.fill_between(x, V(x), 0, alpha=0.2, color='#f55a50')
    plt.plot(x, V(x), color='#f55a50')
    plt.grid()
    plt.show()
    print('\n')

    #Função Momento fletor
    x = np.linspace(0, comprimento)
    def M(x):
        return -((intensidadeForca/2)*x**2) + (intensidadeForca*comprimento*x) - ((1/2)*intensidadeForca*comprimento**2)
    plt.title('Momento Fletor')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('M(x)') #definindo nome do eixo Y
    plt.plot(x, M(x), color='m')
    plt.gca().invert_yaxis()
    plt.fill_between(x, M(x), 0, alpha=0.2, color='m')
    plt.grid()
    plt.show()
    print('\n')

'''Resolvendo o exercicio 3: Força contínua triangular em uma barra engastada'''
def ex3(forcaMax, comprimento):
    #Esforços Reativos
    forcaX = 0
    forcaY = (forcaMax * comprimento) / 2
    momento = (forcaY * comprimento) / 3
    print("Os esforços reativos no engastamento são:")
    print("\nForça horizontal reativa: ", forcaX)
    print("\nForça vertical reativa: ", forcaY)
    print("\nMomento reativo: ", momento)
    print('\n')

    #Função Força Normal
    @np.vectorize
    def N(x):
      return 0
    x = np.linspace(0, comprimento)
    plt.plot(x, N(x), color='xkcd:sky blue')
    plt.title('Normal')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('N(x)') #definindo nome do eixo Y
    plt.grid()
    plt.show()
    print('\n')

    #Função Força Cortante
    def V(x):
      return ((forcaMax * ((comprimento - x) ** 2)) / (2 * comprimento))
    x = np.linspace(0, comprimento)
    plt.plot(x, V(x), color='#f55a50')
    plt.title('Cortante')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('V(x)') #definindo nome do eixo Y
    plt.fill_between(x, V(x), 0, alpha=0.2, color='#f55a50')
    plt.grid()
    plt.show()
    print('\n')
    
    #Função Momento Fletor
    def M(x):
      return ((forcaMax * ( (x - comprimento) ** 3)) / (6 * comprimento))
    x = np.linspace(0, comprimento)
    plt.plot(x, M(x), color='m')
    plt.title('Momento Fletor')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('M(x)') #definindo nome do eixo Y
    plt.gca().invert_yaxis()
    plt.fill_between(x, M(x), 0, alpha=0.2, color='m')
    plt.grid()
    plt.show()
    print('\n')

'''Resolvendo o exercicio 4: Força pontual em uma barra com uma articulação fixa e uma móvel'''
def ex4(intensidadeForca, comprimento):
    pontoDeAplicacao = int(
        input("Defina o ponto de aplicação em relação ao ponto incial: "))
      
    #Esforços reativos
    print("Os esforços reativos no apoio fixo:")
    print("\nForça horizontal reativa: 0")
    print("\nForça vertical reativa: ", intensidadeForca*(comprimento-pontoDeAplicacao)/comprimento)
    print("\nOs esforços reativos no apoio simples:")
    print("\nForça vertical reativa: ", intensidadeForca*pontoDeAplicacao/comprimento)

    #Função Força Normal
    @np.vectorize
    def N(x):
        return 0
    
    plt.title('Força normal')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('N(x)') #definindo nome do eixo Y
    x = np.linspace(0, comprimento)
    plt.plot(x, N(x), color="xkcd:sky blue")
    plt.grid()
    plt.show()

    #Função Força Cortante
    @np.vectorize
    def V1(x1):
        return intensidadeForca * (comprimento -
                                   pontoDeAplicacao) / comprimento

    @np.vectorize
    def V2(x2):
        return -intensidadeForca * pontoDeAplicacao / comprimento

    plt.title('Força Cortante')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('V(x)') #definindo nome do eixo Y
    x1 = np.linspace(0, pontoDeAplicacao)
    x2 = np.linspace(pontoDeAplicacao, comprimento)
    plt.fill_between(x1, V1(x1), 0, alpha=0.2, color='#f55a50')
    plt.fill_between(x2, V2(x2), 0, alpha=0.2, color='#f55a50')
    plt.plot(x1, V1(x1), color='#f55a50')
    plt.plot(x2, V2(x2), color='#f55a50')
    plt.grid()
    plt.show()
    print('\n')

    #Função Momento Fletor
    def M1(x1):
        return intensidadeForca * (comprimento -
                                   pontoDeAplicacao) * x1 / comprimento

    def M2(x2):
        return -intensidadeForca * pontoDeAplicacao * x2 / comprimento + intensidadeForca * pontoDeAplicacao

    plt.title('Momento Fletor')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('M(x)') #definindo nome do eixo Y
    x1 = np.linspace(0, pontoDeAplicacao)
    x2 = np.linspace(pontoDeAplicacao, comprimento)
    plt.gca().invert_yaxis()
    plt.fill_between(x1, M1(x1), 0, alpha=0.2, color='m')
    plt.fill_between(x2, M2(x2), 0, alpha=0.2, color='m')
    plt.plot(x1, M1(x1), color='m')
    plt.plot(x2, M2(x2), color='m')
    plt.grid()
    plt.show()
    print('\n')

'''Resolvendo o exercicio 5: Força contínua retangular em uma barra com uma articulação fixa e uma móvel'''
def ex5(intensidadeForca, comprimento):
    #Esforços Reativos
    #No apoio fixo
    forcaX1 = 0
    forcaY1 = (intensidadeForca * comprimento) / 2
    #No apoio móvel
    forcaY2 = (intensidadeForca * comprimento) / 2

    print("Os esforços reativos no apoio fixo:")
    print("\nForça horizontal reativa: ", forcaX1)
    print("\nForça vertical reativa: ", forcaY1)
    print("\nOs esforços reativos no apoio simples:")
    print("\nForça vertical reativa: ", forcaY2)
    print('\n')
      
    #Função Força Normal
    @np.vectorize
    def N(x):
      return 0
    x = np.linspace(0, comprimento)
    plt.plot(x, N(x), color='xkcd:sky blue')
    plt.title('Normal')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('N(x)') #definindo nome do eixo Y
    plt.grid()
    plt.show()
    print('\n')

    #Função Força Cortante
    def V(x):
      return ((intensidadeForca * comprimento) / 2) - (intensidadeForca * x) 
    x = np.linspace(0, comprimento)
    plt.plot(x, V(x), color='#f55a50')
    plt.title('Cortante')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('V(x)') #definindo nome do eixo Y
    plt.fill_between(x, V(x), 0, alpha=0.2, color='#f55a50')
    plt.grid()
    plt.show()
    print('\n')

    #Função Momento Fletor
    def M(x):
      return ((intensidadeForca * comprimento * x) / 2) - ( (intensidadeForca * (x**2)) / 2)
    x = np.linspace(0, comprimento)
    plt.plot(x, M(x), color='m')
    plt.title('Momento Fletor')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('M(x)') #definindo nome do eixo Y
    plt.gca().invert_yaxis()
    plt.fill_between(x, M(x), 0, alpha=0.2, color='m')
    plt.grid()
    plt.show()
    print('\n')

'''Resolvendo o exercicio 6: Força contínua triangular em uma barra com uma articulação fixa e uma móvel'''
def ex6(forcaMax, comprimento):
  
    #Esforços reativos
    print("Os esforços reativos no apoio fixo:")
    print("\nForça horizontal reativa: 0")
    print("\nForça vertical reativa: ", (forcaMax*comprimento)/6)
    print("\nOs esforços reativos no apoio simples:")
    print("\nForça vertical reativa: ", (forcaMax*comprimento)/3)

    #Função Normal
    @np.vectorize
    def N(x):
        return 0

    t1 = np.linspace(0.0, comprimento)
    plt.plot(t1, N(t1), color='xkcd:sky blue')
    plt.title('Normal')
    plt.grid()
    plt.fill(t1,N(t1),alpha=0.2)
    plt.show()
    print('\n')

    #Função Cortante
    def V(x):
        return ((comprimento*forcaMax)/6) - ((forcaMax/(2*comprimento))*(x**2))
    
    @np.vectorize
    def V1(x):
        return 0

    x = np.linspace(0, comprimento)

    plt.plot(x, V(x), color='#f55a50')
    plt.plot(x, V1(x), color='#f55a50')
    plt.title('Cortante')
    plt.fill_between(x, V(x), 0, alpha=0.2, color='#f55a50')
    plt.grid()
    plt.show()
    print('\n')

    #Função Momento Fletor
    def M(x):
      return (((comprimento*forcaMax)/6)*x) - ((forcaMax/(6*comprimento))*(x**3))

    x = np.linspace(0, comprimento)

    plt.plot(x, M(x), color='m')
    plt.title('Momento Fletor')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('M(x)') #definindo nome do eixo Y
    plt.fill(x,M(x),alpha=0.2, color='m')
    plt.gca().invert_yaxis()
    plt.grid()
    plt.show()

'''Inicio do programa'''
print("Bem-vindo ao programa de PEF3208!\n")
print(
    "Comece digitando as configurações do problema!\nVamos considerar um problema em que o sistema esteja no primeiro quadrante, a barra esteja na horizontal e tenha somente uma força vertical aplicada.\n\n"
)

'''Declaração de Variáveis'''
comprimento = 0
tipoForca = 0
formaContinua = 0
intensidadeForca = 0
densidadeForca = 0
forcaMax = 0
pontoDeAplicacao = 0
modelo = 0

'''Comprimento da barra'''
comprimento = int(
    input("Qual o comprimento da barra? Digite um número inteiro. \n"))
while ((comprimento <= 0)):
    comprimento = int(input("Opção inválida, digite um valor maior que zero:"))
'''Tipo de força da vertical'''
tipoForca = int(
    input(
        "\nQual o tipo de força? Digite:\n1. Força pontual\n2. Força continua\n"
    ))
'''Tipo de força'''
while ((tipoForca != 1) and (tipoForca != 2)):
    tipoForca = int(input("Opção inválida, digite uma opção válida: "))
if (tipoForca == 2):
    formaContinua = int(
        input(
            print(
                "\nDigite a opção que descreva a função de continuidade da força: \n1. Função constante (forma retangular) \n2. Função linear (forma triangular) \n"
            )))

'''Intensidade da Força'''
if (tipoForca == 1):  #forca pontual
    intensidadeForca = int(
        input(
            "\nQual a intensidade da força aplicada? Digite um número inteiro. \n"
        ))
else:  #forca continua
    if (formaContinua == 1):  #forca continua retangular
        densidadeForca = int(
            input(
                "\nQual a função densidade da força aplicada? Digite um número inteiro. \n"
            ))
        intensidadeForca = comprimento * densidadeForca
    else:  #forca continua triangular
        forcaMax = int(
            input(
                "\nQual a força máxima aplicada em um extremo? Digite um número inteiro. \n"
            ))
        intensidadeForca = comprimento * (forcaMax) / 2
        
'''Definindo os vínculos -> Modelo'''
modelo = int(
    input(
        "\nDe acordo com os vínculos do seu modelo, escolha uma opção:\n1. Viga em balanço (com engastamento)\n2. Viga com uma articulação fixa e uma articulação móvel\n"
    ))
while (modelo != 1 and modelo != 2):
    modelo = int(input("Digite um valor válido: "))

if (modelo == 1):
    if (tipoForca == 1):  #ex1
        ex1(intensidadeForca, comprimento)
    else:
        if (formaContinua == 1):  #ex2
            ex2(intensidadeForca, comprimento)
        else:  #ex3
            ex3(forcaMax, comprimento)
else:
    if (tipoForca == 1):  #ex4
        ex4(intensidadeForca, comprimento)
    else:
        if (formaContinua == 1):  #ex5
            ex5(densidadeForca, comprimento)
        else:  #ex6
            ex6(forcaMax, comprimento)
```

```python
def ex4(intensidadeForca, comprimento):
    pontoDeAplicacao = int(
        input("Defina o ponto de aplicação em relação ao ponto incial: "))
      
    #Esforços reativos
    print("Os esforços reativos no apoio fixo:")
    print("\nForça horizontal reativa: 0")
    print("\nForça vertical reativa: ", intensidadeForca*(comprimento-pontoDeAplicacao)/comprimento)
    print("\nOs esforços reativos no apoio simples:")
    print("\nForça vertical reativa: ", intensidadeForca*pontoDeAplicacao/comprimento)
    print('\n')

    #Função Força Normal
    @np.vectorize
    def N(x):
        return 0
    
    plt.title('Força normal')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('N(x)') #definindo nome do eixo Y
    x = np.linspace(0, comprimento)
    plt.plot(x, N(x), color="xkcd:sky blue")
    plt.grid()
    plt.show()

    #Função Força Cortante
    @np.vectorize
    def V1(x1):
        return intensidadeForca * (comprimento -
                                   pontoDeAplicacao) / comprimento

    @np.vectorize
    def V2(x2):
        return -intensidadeForca * pontoDeAplicacao / comprimento

    plt.title('Força Cortante')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('V(x)') #definindo nome do eixo Y
    x1 = np.linspace(0, pontoDeAplicacao)
    x2 = np.linspace(pontoDeAplicacao, comprimento)
    plt.fill_between(x1, V1(x1), 0, alpha=0.2, color='#f55a50')
    plt.fill_between(x2, V2(x2), 0, alpha=0.2, color='#f55a50')
    plt.plot(x1, V1(x1), color='#f55a50')
    plt.plot(x2, V2(x2), color='#f55a50')
    plt.grid()
    plt.show()
    print('\n')

    #Função Momento Fletor
    def M1(x1):
        return intensidadeForca * (comprimento -
                                   pontoDeAplicacao) * x1 / comprimento

    def M2(x2):
        return -intensidadeForca * pontoDeAplicacao * x2 / comprimento + intensidadeForca * pontoDeAplicacao

    plt.title('Momento Fletor')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('M(x)') #definindo nome do eixo Y
    x1 = np.linspace(0, pontoDeAplicacao)
    x2 = np.linspace(pontoDeAplicacao, comprimento)
    plt.gca().invert_yaxis()
    plt.fill_between(x1, M1(x1), 0, alpha=0.2, color='m')
    plt.fill_between(x2, M2(x2), 0, alpha=0.2, color='m')
    plt.plot(x1, M1(x1), color='m')
    plt.plot(x2, M2(x2), color='m')
    plt.grid()
    plt.show()
    print('\n')

ex4(5,10)
```

```python
def ex5(intensidadeForca, comprimento):
    #Esforços Reativos
    #No apoio fixo
    forcaX1 = 0
    forcaY1 = (intensidadeForca * comprimento) / 2
    #No apoio móvel
    forcaY2 = (intensidadeForca * comprimento) / 2

    print("Os esforços reativos no apoio fixo:")
    print("\nForça horizontal reativa: ", forcaX1)
    print("\nForça vertical reativa: ", forcaY1)
    print("\nOs esforços reativos no apoio simples:")
    print("\nForça vertical reativa: ", forcaY2)
    print('\n')
      
    #Função Força Normal
    @np.vectorize
    def N(x):
      return 0
    x = np.linspace(0, comprimento)
    plt.plot(x, N(x), color='xkcd:sky blue')
    plt.title('Normal')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('N(x)') #definindo nome do eixo Y
    plt.grid()
    plt.show()
    print('\n')

    #Função Força Cortante
    def V(x):
      return ((intensidadeForca * comprimento) / 2) - (intensidadeForca * x) 
    x = np.linspace(0, comprimento)
    plt.plot(x, V(x), color='#f55a50')
    plt.title('Cortante')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('V(x)') #definindo nome do eixo Y
    plt.fill_between(x, V(x), 0, alpha=0.2, color='#f55a50')
    plt.grid()
    plt.show()
    print('\n')

    #Função Momento Fletor
    def M(x):
      return ((intensidadeForca * comprimento * x) / 2) - ( (intensidadeForca * (x**2)) / 2)
    x = np.linspace(0, comprimento)
    plt.plot(x, M(x), color='m')
    plt.title('Momento Fletor')
    plt.xlabel('x') #definindo nome do eixo X
    plt.ylabel('M(x)') #definindo nome do eixo Y
    plt.gca().invert_yaxis()
    plt.fill_between(x, M(x), 0, alpha=0.2, color='m')
    plt.grid()
    plt.show()
    print('\n')

ex5(5,10)
```
